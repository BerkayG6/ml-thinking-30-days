# Day 04 — Karar Eşiği (Decision Threshold) ve Precision–Recall Dengesi

## Model Eğitimi Neden Yeterli Değildir?
Makine öğrenmesi projelerinde çok sık yapılan bir hata şudur:
model eğitildiği anda sürecin bittiğinin düşünülmesi.

Oysa özellikle **olasılık (probability) üreten sınıflandırma modellerinde**
(modelimiz Logistic Regression gibi),
eğitim sadece şunu belirler:

> “Bu işlem fraud olma ihtimali açısından kaç puan alıyor?”

Model **doğrudan karar vermez**.
Kararı veren şey, bu olasılığı nasıl yorumladığımızdır.

---

## Olasılık Tahmini ile Sınıf Kararı Arasındaki Fark
Bir sınıflandırma modeli şunu üretir:

- Bu işlem için `P(fraud) = 0.73`
- Başka bir işlem için `P(fraud) = 0.12`

Ancak model şunu söylemez:
> “Bu kesin fraud”  
> “Bu kesin değil”

Bu noktada devreye **karar eşiği (decision threshold)** girer.

---

## Karar Eşiği (Decision Threshold) Nedir?
Karar eşiği, modelin ürettiği olasılığı
**sınıf etiketine dönüştürmek için kullanılan sınırdır**.

Varsayılan kural şudur:
- Eğer `P(fraud) ≥ 0.5` → fraud
- Eğer `P(fraud) < 0.5` → fraud değil

Ama burada çok önemli bir gerçek var:

> **0.5 evrensel, doğru veya zorunlu bir eşik değildir.**

Bu değer:
- Veri setinden gelmez
- İş probleminden gelmez
- Sadece matematiksel bir varsayımdır

---

## Neden 0.5 Eşiği Fraud Problemi İçin Yanlıştır?
Fraud detection problemleri **aşırı dengesizdir**:
- Fraud işlemler çok azdır
- Normal işlemler çok fazladır

Bu durumda:
- Modelin “fraud” demesi zaten zor
- 0.5 gibi yüksek bir eşik, birçok fraud’un kaçmasına neden olabilir

Üstelik:
- Fraud kaçırmanın (False Negative) maliyeti çok yüksektir
- Yanlış alarmın (False Positive) maliyeti görece düşüktür

Bu yüzden:
> **Karar eşiği, iş maliyetine göre ayarlanmalıdır.**

---

## Precision ve Recall Neden Birbirine Zıttır?
Karar eşiğini değiştirdiğimizde şunlar olur:

### Eşik Düşürüldüğünde (örneğin 0.5 → 0.3)
- Daha fazla işlem “fraud” olarak işaretlenir
- **Recall artar** (daha az fraud kaçırılır)
- **Precision düşer** (daha fazla yanlış alarm oluşur)

### Eşik Yükseltildiğinde (örneğin 0.5 → 0.7)
- Daha az işlem “fraud” olarak işaretlenir
- **Precision artar**
- **Recall düşer** (daha çok fraud kaçırılır)

Bu yüzden precision ve recall arasında doğal bir **denge (trade-off)** vardır.

---

## Fraud Detection Perspektifinden Bakış
Fraud detection’da temel soru şudur:

> “Bir fraud’u kaçırmak mı daha pahalı,  
> yoksa birkaç normal işlemi yanlışlıkla durdurmak mı?”

Genellikle cevap nettir:
- Fraud kaçırmak daha pahalıdır

Bu yüzden:
- Recall genellikle önceliklidir
- Precision ikinci plandadır

Ancak bu, precision’ın önemsiz olduğu anlamına gelmez.
Bu denge **bilinçli şekilde yönetilmelidir**.

---

## Threshold Ayarlamak Ne Kazandırır?
Threshold tuning sayesinde:
- Aynı modeli yeniden eğitmeden
- Model davranışını kontrol edebiliriz

Yani:
> **Model sabit kalır, karar stratejisi değişir.**

Bu yaklaşım:
- Daha şeffaf
- Daha açıklanabilir
- Daha iş odaklıdır

---

## Day 04’ün Ana Mesajı
Day 04’te öğrenilmesi gereken temel fikir şudur:

> **Makine öğrenmesi modelleri olasılık üretir,  
> iş dünyası karar ister.  
> Karar eşiği, bu ikisi arasındaki köprüdür.**

Bu köprü doğru kurulmazsa:
- En iyi model bile yanlış kararlar verir.
