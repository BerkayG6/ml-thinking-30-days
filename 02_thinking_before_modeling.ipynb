# GÜN 2 — Modelden Önce Düşünmek: Feature’lara Doğru Bakış

## Giriş

Makine öğrenmesine yeni başlayanların büyük bir kısmı veri setini açtığı anda
şu refleksi gösterir:

> “Önce bir `df.describe()` yapayım,  
> sonra birkaç grafik çizeyim,  
> sonra da model deneyelim.”

Bu yaklaşım kısa vadede ilerleme hissi verse de,
uzun vadede hatalı modellere, yanlış yorumlara
ve en önemlisi **gerçek hayatta çalışmayan çözümlere** yol açar.

GÜN 2’nin amacı,
bu refleksi bilinçli bir şekilde durdurmak
ve modelleme öncesi **düşünme disiplinini** kazandırmaktır.

---

## Feature Nedir? (Yanlış Bilinen Bir Kavram)

Birçok kişi feature’ı şu şekilde düşünür:
> “Modelin kullandığı sütun.”

Bu tanım teknik olarak doğru olsa da **eksiktir**.

Daha doğru tanım şudur:

> **Feature, modelin karar verirken kullandığı bilgidir.**

Bu nedenle her feature üç açıdan değerlendirilmelidir:
1. Bilgi içeriyor mu?
2. Tahmin anında mevcut mu?
3. Yanlış bir avantaj sağlıyor olabilir mi?

GÜN 2’nin temel sorusu tam olarak budur.

---

## Feature’lara Anlamsal Değil, Fonksiyonel Bakmak

Özellikle anonimleştirilmiş veri setlerinde (örneğin PCA uygulanmış fraud dataset’lerinde)
feature’ların anlamsal karşılıkları bilinmez.

Bu bir dezavantaj değildir.

Çünkü gerçek hayatta da birçok model:
- sinyal,
- frekans,
- istatistiksel davranış

üzerinden çalışır.

Bu yüzden GÜN 2’de feature’lara şu gözle bakılır:

> “Bu feature neyi temsil ediyor?” değil,  
> **“Bu feature modelin hangi kararını etkiler?”**

---

## Feature Sınıflandırması: Amaç Değil, Araç

Feature’ları gruplara ayırmak (target, zaman bilgisi, işlem özelliği vb.)
tek başına bir hedef değildir.

Bu sınıflandırma:
- Leakage risklerini erken fark etmek,
- EDA’da gereksiz grafiklerden kaçınmak,
- Modelleme öncesi doğru varsayımlar kurmak

için kullanılan **bir araçtır**.

Bu nedenle GÜN 2’de yapılan sınıflandırma,
bir “listeleme” değil,
bir **risk analizi**dir.

---

## Tahmin Anında Mevcut Olma İlkesi

GÜN 2’nin en kritik prensibi şudur:

> **Bir feature, tahmin anında mevcut değilse,
> o feature potansiyel olarak tehlikelidir.**

Bu ilke şunları kapsar:
- Geleceğe ait bilgiler
- Olaydan sonra oluşan durumlar
- Toplu (aggregate) istatistikler

Bir feature teknik olarak doğru hesaplanmış olabilir,
ama gerçek hayatta tahmin anında bilinmiyorsa,
modeli yanıltır.

---

## Ölçek (Scaling) Bir Görsellik Meselesi Değildir

Yeni başlayanlar için scaling genellikle
“model daha iyi çalışsın” gibi algılanır.

Oysa GÜN 2 bakış açısında scaling:
> **Feature’lar arasında adaleti sağlama aracıdır.**

Örneğin:
- PCA ile üretilmiş feature’lar zaten ölçeklidir
- Zaman bilgisi sıralama taşır, büyüklük değil
- Amount gibi değişkenler modeli domine edebilir

Bu nedenle scaling kararı,
model kurulduktan sonra değil,
**feature incelenirken** verilmelidir.

---

## Neden EDA’ya Hemen Başlamıyoruz?

EDA (Exploratory Data Analysis),
doğru sorularla yapılmadığında
grafik gürültüsüne dönüşür.

GÜN 2’nin temel yaklaşımı şudur:

> Grafik çizmeden önce,
> “Bu grafik bana hangi kararı verdirecek?”
> sorusu sorulmalıdır.

Aksi halde:
- Anlamsız korelasyonlar
- Yanlış çıkarımlar
- Modeli yönlendiren hatalı sezgiler
oluşur.

GÜN 2, EDA’yı yasaklamaz;
EDA’yı **amaçlı hale getirir**.

---

## GÜN 2’nin Asıl Kazandırdığı Şey

GÜN 2 sonunda kazanılan şey:
- Yeni bir algoritma bilgisi değildir
- Yeni bir kütüphane değildir

GÜN 2’nin kazandırdığı şey şudur:

> **Model kurmadan önce düşünen data scientist refleksi**

Bu refleks kazanıldığında:
- Dataset değişse bile yaklaşım değişmez
- Algoritma değişse bile karar mantığı korunur
- Model başarısız olduğunda nedenini açıklamak mümkün olur

---

## Sonuç

GÜN 2,
makine öğrenmesi sürecinin en sessiz ama en kritik günüdür.

Bu gün atlanırsa:
- Model kurulur
- Sonuç alınır
- Ama neden çalıştığı veya çalışmadığı bilinmez

Bu gün doğru yapılırsa:
- Daha az model denenir
- Daha doğru kararlar alınır
- Gerçek hayatta çalışan sistemler kurulur

Bu nedenle GÜN 2,
bir teknik gün değil,
**bir düşünme günüdür**.
